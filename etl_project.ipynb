```python
import pandas as pd
from sqlalchemy import create_engine
```

## Extract CSVs into DataFrames, clean data and create new data with select columns

```python
COVIDDeathsByState_file = "COVIDDeathsByState.csv"
COVIDDeathsByState_df = pd.read_csv(COVIDDeathsByState_file)
COVIDDeathsByState_df.head()
```

```python
COVIDDeathsByState_columns = ["submission_date", "state", "tot_death"]
COVIDDeathsByState_df_small = COVIDDeathsByState_df[COVIDDeathsByState_columns].copy()
COVIDDeathsByState_df_small.head()
```

```python
COVIDDeathsByState_ColumnsRenamed = COVIDDeathsByState_df_small.rename(columns={"submission_date": "date","state": "state",
"tot_death": "totalcoviddeaths"})
COVIDDeathsByState_ColumnsRenamed.set_index('date', inplace = True)
COVIDDeathsByState_ColumnsRenamed.head()
```

```python
Population_file = "Population.csv"
Population_df = pd.read_csv(Population_file)
Population_df.head()
```

```python
Population_columns = ["State", "Population", "Hospitals"]
Population_df_small = Population_df[Population_columns].copy()
Population_df_small.head()
```

```python
Population_ColumnsRenamed = Population_df_small.rename(columns={"State": "state","Population": "population",
"Hospitals": "hospitals"})
Population_ColumnsRenamed.set_index('state', inplace = True)
Population_ColumnsRenamed.head()
```

```python
TotalDeaths_file = "TotalDeaths.csv"
TotalDeaths_df = pd.read_csv(TotalDeaths_file)
TotalDeaths_df.head()
```

```python
TotalDeaths_columns = ["State", "Year", "Month", "Indicator", "Data Value"]
TotalDeaths_df_small = TotalDeaths_df[TotalDeaths_columns].copy()
TotalDeaths_df_small.head()
```

```python
TotalDeaths_ColumnsRenamed = TotalDeaths_df_small.rename(columns={"State": "state","Year": "yr","Month": "mth",
"Indicator": "indicator","Data Value": "data_value"})
TotalDeaths_ColumnsRenamed.set_index('state', inplace = True)
TotalDeaths_ColumnsRenamed.head()
```

## Create database connection

```python
rds_connection_string = "postgres:postgres@localhost:5432/CovidDeathFigures"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

## Check for tables

```python
engine.table_names()
```

## Use pandas to load CSV converted DataFrames into database

```python
COVIDDeathsByState_ColumnsRenamed.to_sql(name='coviddeathsbystate', con=engine, if_exists='append', index=True)
```

```python
Population_ColumnsRenamed.to_sql(name='population', con=engine, if_exists='append', index=True)
```

```python
TotalDeaths_ColumnsRenamed.to_sql(name='totaldeaths', con=engine, if_exists='append', index=True)
```

## Confirm data has been added by querying the tables

```python
pd.read_sql_query('select * from coviddeathsbystate', con=engine).head()
```

```python
pd.read_sql_query('select * from population', con=engine).head()
```

```python
pd.read_sql_query('select * from totaldeaths', con=engine).head()
```
